# Advanced Querying Mongo

Importing libraries and setting up connection

```python
%pip install pymongo
from pymongo import MongoClient
client = MongoClient("mongodb://localhost:27017")
import pymongo
import pandas as pd
```

### 1. All the companies whose name match 'Babelgum'. Retrieve only their `name` field.

```python
db = client["companies"]
collection = db["collection"]

query = {'name':'Babelgum'}
projection = {'name': 1, '_id': 0}
result = collection.find(query, projection)

list(result)
                  
```

### 2. All the companies that have more than 5000 employees. Limit the search to 20 companies and sort them by **number of employees**.

```python
query = {'number_of_employees': {'$gt': 5000}}
res = collection.find(query).sort('number_of_employees', -1).limit(20)
pd.DataFrame(res)
```

### 3. All the companies founded between 2000 and 2005, both years included. Retrieve only the `name` and `founded_year` fields.

```python
query = {
    'founded_year': {
        '$gte': 2000,  
        '$lte': 2005   
    }
}

projection = {'name': 1, 'founded_year': 1, '_id': 0}   
result = collection.find(query, projection).limit(10)
list(result)
```

### 4. All the companies that had a Valuation Amount of more than 100.000.000 and have been founded before 2010. Retrieve only the `name` and `ipo` fields.

```python
query = {
    '$and': [
        {'ipo.valuation_amount': {'$gte': 100000000}},
        {'founded_year': {'$lte': 2010}}
    ]
}

projection = {'name': 1, 'ipo.valuation_amount': 1, '_id': 0}   
result = collection.find(query, projection).limit(10)
list(result)
```

### 5. All the companies that have less than 1000 employees and have been founded before 2005. Order them by the number of employees and limit the search to 10 companies.

```python
query = {
    '$and': [
        {'number_of_employees': {'$lte': 1000}},
        {'founded_year': {'$lte': 2005}}
    ]
}

projection = {'name': 1, 'number_of_employees': 1,'founded_year':1, '_id': 0}   
result = collection.find(query, projection).sort('number_of_employees', -1).limit(10)
list(result)
```

### 6. All the companies that don't include the `partners` field.

```python
query = {'partners':{'$exists':0}}
projection = {'name': 1, '_id': 0}   
result = collection.find(query, projection).limit(10)    
list(result)
```

### 7. All the companies that have a null type of value on the `category_code` field.

```python
query = {'category_code': None}
projection = {'name': 1, '_id': 0}
result = collection.find(query, projection).limit(10)    
list(result)
```

### 8. All the companies that have at least 100 employees but less than 1000. Retrieve only the `name` and `number of employees` fields.

```python
query = {
    '$and': [
        {'number_of_employees': {'$gte': 100}},
        {'number_of_employees': {'$lte': 1000}}
    ]
}
projection = {'name': 1,'number_of_employees': 1, '_id': 0}
res = collection.find(query,projection).sort('number_of_employees', -1).limit(20)
list(res)
```

### 9. Order all the companies by their IPO price in a descending order.

### 11. All the companies founded on the second semester of the year. Limit your search to 1000 companies.

```python
query = {
    'founded_month': {
        '$gte': 6,  
        '$lte': 12   
    }
}

projection = {'name': 1, 'founded_month': 1, '_id': 0}   
result = collection.find(query, projection).limit(10)
list(result)
```

### 12. All the companies founded before 2000 that have an acquisition amount of more than 10.000.00

```python
query = {
    '$and': [
        {'founded_year': {'$lte': 2000}},
        {'acquisition.price_amount': {'$gte': 100000}}
    ]
}
projection = {'name': 1,'founded_year': 1,'acquisition.price_amount':1, '_id': 0}
res = collection.find(query,projection).sort('acquisition.price_amount', -1).limit(10)
list(res)
```

### 13. All the companies that have been acquired after 2010, order by the acquisition amount, and retrieve only their `name` and `acquisition` field.

```python
query = {'acquisition.acquired_year': {'$gt':2010}}
projection = {'name': 1,'acquisition.price_amount':1, '_id': 0}
res = collection.find(query,projection).sort('acquisition.price_amount', -1).limit(10)
list(res)
```

### 14. Order the companies by their `founded year`, retrieving only their `name` and `founded year`.

### 15. All the companies that have been founded on the first seven days of the month, including the seventh. Sort them by their `acquisition price` in a descending order. Limit the search to 10 documents.

```python
query = {'founded_day': {'$lte':7}}
projection = {'name': 1, 'founded_day': 1, 'acquisition.price_amount':1, '_id': 0}   
result = collection.find(query, projection).sort('acquistion.price_amount',-1).limit(10)
list(result)
```

### 16. All the companies on the 'web' `category` that have more than 4000 employees. Sort them by the amount of employees in ascending order.

```python
query = {'$and': [{'category_code': 'web'} , {'number_of_employees': {'$gt':4000}}]}
projection = {'name': 1, 'category_code': 1, 'number_of_employees':1, '_id': 0}   
result = collection.find(query, projection).sort('number_of_employees',1).limit(10)
list(result)
```

### 17. All the companies whose acquisition amount is more than 10.000.000, and currency is 'EUR'.

```python
query = {'$and': [{'acquisition.price_currency_code': 'EUR'},
                {'acquisition.price_amount': {'$gt': 10000000}}]}
select = {'_id': 0, 'name': 1, 'acquisition.price_amount':1}
list(collection.find(query, select).limit(10))
```

### 18. All the companies that have been acquired on the first trimester of the year. Limit the search to 10 companies, and retrieve only their `name` and `acquisition` fields.

```python
query = {
    "acquisition.acquired_month": {
        '$gte': 1,  
        '$lte': 6   
    }
}

projection = {'name': 1, 'acquisition.acquired_month': 1, '_id': 0}   
result = collection.find(query, projection).limit(10)
list(result)
```
